```python
# ==============================================================
# BACKEND WORDPRESS DINÀMIC - AMB SEGON TÒKEN NGROK
# ==============================================================

# 1. INSTAL·LACIÓ DE LLIBRERIES
!pip install -U google-genai flask-cors pyngrok beautifulsoup4 requests

import os, time, json, requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse
from flask import Flask, request, jsonify
from flask_cors import CORS
from pyngrok import ngrok
from google import genai
from google.colab import userdata

# ==============================================================
# 2. CONFIGURACIÓ DE CREDENCIALS I URL
# ==============================================================
# La teva API KEY de Gemini (segueix estant als Secrets 🔑)
GOOGLE_API_KEY = userdata.get("GOOGLE_API_KEY")

# EL TEU NOU TÒKEN DE NGROK (Posat directament com has demanat)
NGROK_TOKEN = "3CLNZLE7Eh663CxpirvcbLV8Yxg_2bphjRyybtPXb4qJVRZXo"

# URL del teu WordPress de l'institut
URL_WORDPRESS = "https://mespinosa.inscastellbisbal.net/"

# Client oficial de Gemini
client = genai.Client(api_key=GOOGLE_API_KEY)

# Base de dades on es guardarà el text del teu WordPress
base_de_dades = []

# ==============================================================
# 3. EXTRACTOR (SCRAPER) - LLEGEIX EL TEU WORDPRESS
# ==============================================================
def escanejar_wordpress(url_base):
    global base_de_dades
    print(f"🚀 Iniciant extracció de dades a: {url_base}")
    urls_visitades = set()
    urls_per_visitar = [url_base]
    domini = urlparse(url_base).netloc

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
    }

    while urls_per_visitar and len(urls_visitades) < 60:
        url = urls_per_visitar.pop(0)

        if url in urls_visitades or any(x in url for x in ["wp-admin", "wp-login", "feed", "comments"]):
            continue

        try:
            # Timeout de 15 segons per evitar els errors de "Read timed out"
            res = requests.get(url, headers=headers, timeout=15)
            if res.status_code != 200: continue

            soup = BeautifulSoup(res.text, 'html.parser')
            urls_visitades.add(url)

            titol = soup.title.string.replace(" – Home", "").strip() if soup.title else url

            # Busquem el contingut de la pàgina
            contingut_html = soup.find('article') or soup.find('main') or soup.body
            parrafs = contingut_html.find_all(['p', 'h1', 'h2', 'h3', 'li'])
            text_paguina = " ".join([p.get_text(strip=True) for p in parrafs])

            if len(text_paguina) > 100:
                base_de_dades.append({"titol": titol, "text": text_paguina})
                print(f"✅ Escanejada OK: {titol[:50]}...")

            # Buscar nous enllaços
            for a in soup.find_all('a', href=True):
                enllac = urljoin(url_base, a['href'])
                enllac = enllac.split('#')[0].split('?')[0]
                if urlparse(enllac).netloc == domini and enllac not in urls_visitades:
                    urls_per_visitar.append(enllac)

            time.sleep(0.5) # Pausa per seguretat

        except Exception as e:
            print(f"⏳ Saltant {url} per lentitud.")

    print(f"\n✅ EXTRACCIÓ COMPLETADA: {len(base_de_dades)} pàgines a la memòria.")

# ==============================================================
# 4. SERVIDOR FLASK (CONEXIÓ AMB EL FRONT-END)
# ==============================================================
app = Flask(__name__)
CORS(app)

@app.route('/ask', methods=['POST', 'OPTIONS'])
def ask():
    if request.method == 'OPTIONS': return jsonify({"status": "ok"}), 200

    try:
        data = request.json
        pregunta = data.get("message", "")

        # Filtre de context intel·ligent
        context_rellevant = ""
        for pagina in base_de_dades:
            if any(paraula.lower() in pagina['text'].lower() for paraula in pregunta.split() if len(paraula) > 3):
                context_rellevant += f"\n--- INFO DE: {pagina['titol']} ---\n{pagina['text'][:900]}\n"

        system_prompt = f"""
        Ets l'assistent personal de Martí Espinosa.
        Respon en català basant-te en la informació del seu WordPress:
        {context_rellevant}
        """

        resposta = client.models.generate_content(
            model="gemini-2.5-flash",
            contents=f"{system_prompt}\n\nPregunta de l'usuari: {pregunta}"
        )

        return jsonify({"reply": resposta.text})

    except Exception as e:
        return jsonify({"reply": "Ho sento, hi ha un error al backend."}), 500

# ==============================================================
# 5. EXECUCIÓ
# ==============================================================
if __name__ == '__main__':
    # Pas 1: Llegir el WordPress
    escanejar_wordpress(URL_WORDPRESS)

    # Pas 2: Netejar ports i activar Ngrok
    os.system("fuser -k 5000/tcp")
    time.sleep(1)
    ngrok.kill()

    # Configurem el tòken que m'has passat
    ngrok.set_auth_token(NGROK_TOKEN)
    public_url = ngrok.connect(5000).public_url

    print(f"\n" + "="*60)
    print(f"🤖 ASSISTENT WORDPRESS ACTIU (Amb tòken nou)")
    print(f"🌐 URL PER AL TEU FRONT-END: {public_url}/ask")
    print("="*60 + "\n")

    app.run(port=5000, host='0.0.0.0', debug=False, use_reloader=False)
```
